15 Write a program to reverse all sublists in a nested list.
Input: [[1, 2, 3], [4, 5, 6], [7, 8, 9]]


```python
li=[[1, 2, 3], [4, 5, 6], [7, 8, 9]]
reversed_list=[]
for i in li:
  reversed_list.append(i[::-1])
print(reversed_list)
```
